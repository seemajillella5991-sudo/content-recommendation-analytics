```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
genres = ['Action', 'Comedy', 'Drama', 'Thriller', 'Romance', 'Sci-Fi', 'Documentary']

content_list = []

for i in range(1, 201):
    content = {
        'content_id': f'C{i:03}',
        'title': f'Title_{i}',
        'genre': random.choice(genres),
        'type': random.choice(['Movie', 'Show']),
        'release_year': random.randint(2000, 2024)
    }
    content_list.append(content)

content_catalog = pd.DataFrame(content_list)
content_catalog.head()
```

```python
watch_history = []

for user in range(1, 301):  # 300 users
    for _ in range(random.randint(20, 60)):  # each user watches 20-60 contents
        watch = {
            'user_id': f'U{user:04}',
            'content_id': random.choice(content_catalog['content_id'].values),
            'watch_time_min': random.randint(5, 180),
            'rating': random.randint(1, 5),
            'watch_date': datetime.now() - timedelta(days=random.randint(1, 365))
        }
        watch_history.append(watch)

user_watch_history = pd.DataFrame(watch_history)
user_watch_history.head()
```

```python
content_catalog.to_csv('../data/content_catalog.csv', index=False)
user_watch_history.to_csv('../data/user_watch_history.csv', index=False)
```
